```python
# Translate Russian into English
headers = ['region', 'city', 'parent_category_name', 'category_name', 'param_1', 'param_2', 'param_3', 'description', 'title',  
           'region_translation', 'city_translation', 'parent_category_name_translation', 'category_name_translation','param_1_translation', 'param_2_translation', 'param_3_translation', 'description_translation', 'title_translation']
#headers = ['region', 'city', 'region_translation', 'city_translation']
translator = Translator()
# Init empty dataframe with much rows as `data`
df = pd.DataFrame(index=range(0,len(temp)), columns=headers)

def translate_row(row):
    ''' Translate elements A and B within `row`. '''
    if isinstance (row[2], str):
        a = translator.translate(row[2], dest='En')
    else:
        a = translator.translate(str(row[2]), dest='En')
    if isinstance (row[3], str):
        b = translator.translate(row[3], dest='En')
    else:
        b = translator.translate(str(row[3]), dest='En')
    if isinstance (row[4], str):
        c = translator.translate(row[4], dest='En')
    else:
        c = translator.translate(str(row[4]), dest='En')
    if isinstance (row[5], str):
        d = translator.translate(row[5], dest='En')
    else:
        d = translator.translate(str(row[5]), dest='En')
    if isinstance (row[6], str):
        e = translator.translate(row[6], dest='En')
    else:
        e = translator.translate(str(row[6]), dest='En')
    if isinstance (row[7], str):
        f = translator.translate(row[7], dest='En')
    else:
        f = translator.translate(str(row[7]), dest='En')  
    if isinstance (row[8], str):
        g = translator.translate(row[8], dest='En')
    else:
        g = translator.translate(str(row[8]), dest='En')  
    if isinstance (row[9], str):
        h = translator.translate(row[9], dest='En')
    else:
        h = translator.translate(str(row[9]), dest='En')
    if isinstance (row[10], str):
        i = translator.translate(row[10], dest='En')
    else:
        i = translator.translate(str(row[10]), dest='En')
        
    return pd.Series([a.origin, b.origin, c.origin, d.origin, e.origin, f.origin, g.origin, h.origin, i.origin,
                      a.text, b.text, c.text, d.text, e.text, f.text, g.text, h.text, i.text], headers)

count = 0
for i, row in enumerate(temp.values):
    # Fill empty dataframe with given serie.
    df.loc[i] = translate_row(row)
    count = count + 1

#print(df)
```
